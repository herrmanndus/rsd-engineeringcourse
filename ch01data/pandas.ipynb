```python
import numpy as np # highly recommended to always load both. some pd functionality requries numpy, such as df.describe(include=np.number)
import pandas as pd
pd.set_option("display.precision", 2)
```

```python
# read some data with pandas read function 
df = pd.read_csv("boston_house_prices.csv",header=1) # https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html
#read_json()
#read_html()
#read_sql()
df.head()
```

```python
#examine the DataFrame more closely 
print(type(df))
print(df.shape) # 506 instances of 14 features 

# this is similar to the struct objects in Matlab
```

```python
# check out the features (columns)
df.columns # no ()
```

```python
# get general info 
print(df.info())
```

```python
# convert between data formats for a specific column
df['CHAS'] = df['CHAS'].astype('float64') # overwrite columns with column newly formatted as float64
```

```python
# THIS IS SUPER USEFUL 
# for each feature, compute mean, median, sd, ... 

df.describe()
```

```python
# describe only objects with a specific data type

df.describe(include=["object", "int64"])
```

```python
df.describe(include=np.number)
# include numeric columns only

#df.describe(include=['object'])
# include str objects

df.describe(include=['int64'])
# include int64 objects only

df.describe(include=['float64'])
# include float64 objects only
```

```python
# count occurences of a particular value in an object
# makes most sense for bool, but works for int

# syntax: df["column"].value_counts() and df["column"].value_counts(normalize = True)
occurences = df["RAD"].value_counts()
proportions = df["RAD"].value_counts(normalize=True)

print(occurences)
print(proportions)
```

```python
# sort by a specific column in ascending or descending order
# syntax: df.sort_values(by='column', ascending=True/False)

df.sort_values(by='TAX', ascending=False)#.head()
```

```python
# sort by multiple columns at once!!!

df.sort_values(by=['TAX','DIS'], 
               ascending=[False,True])

#NOTE: have to supply multiple 'ascending' attributes as well! 
```

```python
# indexing and basic stats

df['INDUS'].mean()
```

```python
# conditional indexing, where TAX > 400, get the means of all other features
# syntax: df[df['column'] (>/</==) x].mean()/.max().min()

df[df['TAX'] > 400].mean()
```

```python
# can also combine with column indexing 
# syntax: df[df['column1'] > x]['column2'].operation()
df[df['TAX'] > 400]['RM'].max()
```

```python
# multiple conditional indexing:
# INDUS > 20 & AGE < 50

# IMPORTANT: note the () around each condition!!!
df[(df['INDUS']>20) & (df['AGE']>50)]['RM'].min()

# syntax: df[(df['column1' > x]) & (df[df['column2']< y])]['index_column'].operation()
```

# complex indexing

```python
# can index as above by columns (features) or indices

# loc() method: index by NAME
# iloc() method: index by NUMBER (similar to numpy.ix_)

# NOTE: loc still indexes ROWS by index (because they don't have a name...)
```

```python
## loc() method 

# syntax: df.loc(row from:row to, 'column from':'column to')

df.loc[0:5, 'ZN':'RAD'] #.max()/other operations perfectly possible

# two important notes: 
# 1: range is INCLUSIVE of top (python atypical)
# 2: loc uses [] to denote the slice ranges
```

```python
## iloc() method 

# syntax: df.iloc[0:5,0:5]

df.iloc[0:5,0:5]

# CRITICAL!!!
```
