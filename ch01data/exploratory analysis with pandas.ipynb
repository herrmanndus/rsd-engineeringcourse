# complex indexing

```python
# can index as above by columns (features) or indices

# loc() method: index by NAME
# iloc() method: index by NUMBER (similar to numpy.ix_)

# NOTE: loc still indexes ROWS by index (because they don't have a name...)
```

```python
## loc() method 

# syntax: df.loc(row from:row to, 'column from':'column to')

df.loc[0:5, 'ZN':'RAD'] #.max()/other operations perfectly possible

# two important notes: 
# 1: range is INCLUSIVE of top (python atypical)
# 2: loc uses [] to denote the slice ranges
```

```python
## iloc() method 

# syntax: df.iloc[0:5,0:5]

df.iloc[0:5,0:5]

# CRITICAL!
# upper range is noninclusive
# 0:5 is first 5 rows/columns (0,1,2,3,4)
```

```python
# some special indexing cases
# first line of dataFrame
df[:1]

# last line of dataFrame
df[-1:] # having nothing after the : means 'to end'

# could do same with iloc...

df.iloc[0:1,]
df.iloc[-1:,]
```

# Mapping values in DataFrames

```python
# replacing values in a column by passing a dict
# defining old and new values:

# syntax (2 lines): 
# condition = {old_value:new_value}
# df['column'] = df['column'].map(condition)

# syntax (1 line):
# df['column'] = df['column'].map({old_value:new_value})
# inspect: df.loc[:,'AGE']

df["AGE"] = df["AGE"].map({61.2:61.1})
df[df["AGE"] == 61.2]
```

```python
# the .replace() method does the exact same thing
# but with a different syntax
# syntax:
#df = df.replace({'column':{before:after}})

df['PTRATIO']
df = df.replace({'PTRATIO':{15.3:13}})
df['PTRATIO']

# note the double dictionary style
# this is very useful if you want to replace multiple values in multiple feature columns
```

# Grouping
summarize or apply some statistics after splitting the data based on some variable 

```python
# works best with bool arguments or discrete values 
# to reformat the table, split DataFrame based on colum of interest values

# set all AGE > 66 to 1 and <+ 66 to 0
df.loc[df['AGE'] <= 70,'AGE'] = 0
df.loc[df['AGE'] > 70,'AGE'] = 1
```

```python
# syntax of grouping function
#df.groupby(by=grouping_columns)[columns_to_show].function(

df.groupby(by='AGE')['TAX'].mean()
```

```python
df.groupby(by='AGE')['TAX','PTRATIO'].mean()
```

```python
# exercise, group by 'RAD'

df.groupby(by='RAD').max()
```

```python
# you can select a list of function to apply to the grouped datasets
# super useful for summary stats depending on grouped stats

# demo: 
# split PTRATIO into three bins
# compute min, mean, max on the groups subsets 

df.loc[df['PTRATIO'] <= 16,'PTRATIO'] = 0
df.loc[(df['PTRATIO'] > 16) & (df['PTRATIO'] <= 19),'PTRATIO'] = 1
df.loc[(df['PTRATIO'] > 19),'PTRATIO'] = 2
```

```python
df.groupby(by='PTRATIO').agg([np.mean,np.std,np.min,np.max])
```

```python
# repeat by only use specific columns (AGE,TAX) and feed a custom function 
inds = ['AGE','TAX']
#print(type(inds))

def grouped_fct(l_in):
    #print(l_in)
    return np.std(l_in)+np.max(l_in)

df.groupby(by='PTRATIO')[inds].agg([np.mean,np.std,np.min,grouped_fct]) # aggregated 
```

# Summary tables

```python
# use the pd.crosstab() function. note: feed columns into this function, can NOT just say df.crosstab!!!

pd.crosstab(df['AGE'],df['PTRATIO'])
```

```python
# normalize to get a lookup table of percentages (occupancy table)

crosstab = pd.crosstab(df['AGE'],df['PTRATIO'], normalize=True)
```

```python
# use np.asarray to transfer a pd DataFrame into a numpy array --> perform matrix computations on it 
t2 = np.asarray(crosstab)
print(t2)
print(np.sum(t2))
# as expected the normalized crosstab entried sum to 1
```

```python
# IMPORTANT NOTE:
# you can also make crosstables from combined binary features! 
df.crosstab(df['feature A'] & df['feature B'],df['feature C'])
```

# Pivot tables

```python
# compute summary statistics (defined value columns) after grouping based on a grouping feature
df.pivot_table(
    ['TAX','CRIM'],
    ['AGE','PTRATIO'],
    aggfunc=["mean","max"]
)
```

```python
# this seems to be similar to group_by()

df.groupby(['AGE','PTRATIO'])[['CRIM','TAX']].agg([np.mean,np.max])
# but note the different notation of .agg functions 
# for pivot_table it's "non-numpy functions", here we have np.operations (without the " ")
```

```python
# add a new column with computed data
# e.g. multiply PTRATIO with AGE and add this as a new column to the end of the df

p_times_a = df['PTRATIO']*df['AGE']

# the insertion syntax is:
# df.insert(loc=location, column='new name', value=new_data)
# the lod = insert locations describes the column AFTER which to insert the feature entry

df.insert(loc=len(df.columns), column = 'PTRATIO times AGE', value = p_times_a)
```
